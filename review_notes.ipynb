A **segment** $\sigma$ is defined as a sequence of states and actions $\{s_k, a_k, \ldots, s_{k+H}, a_{k+H}\}$.




**Bradley-Terry model**

The probability that trajectory segment $\sigma_1$ is preferred over $\sigma_0$ is given by:

$$
P_\psi [\sigma_1 \succ \sigma_0] = \frac{\exp\left(\sum_t \hat{r}_\psi(s^1_t, a^1_t)\right)}{\sum_{i \in \{0,1\}} \exp\left(\sum_t \hat{r}_\psi(s^i_t, a^i_t)\right)}
$$

It comes from preference-based reinforcement learning or reward learning from human preferences, particularly as used in preference-based reward modeling such as in [Christiano et al., 2017].

> The probability, according to the reward model $\psi$, that trajectory $\sigma_1$ is preferred over trajectory $\sigma_0$, is equal to the exponential of the total predicted reward for $\sigma_1$, divided by the sum of exponentials of the total predicted rewards for both $\sigma_0$ and $\sigma_1$.

---

#### What the terms mean:
- $\sigma_0$, $\sigma_1$: Two trajectory segments (i.e., short sequences of state-action pairs).
- $\sigma_1 \succ \sigma_0$: Human prefers $\sigma_1$ over $\sigma_0$.
- $\hat{r}_\psi(s_t, a_t)$: A learned reward function parameterized by $\psi$, estimating the reward at time $t$.
- The numerator computes the **exponentiated cumulative reward** for $\sigma_1$.
- The denominator normalizes over both trajectories to get a probability distribution.

#### Intuition:
- If $\sigma_1$ has a higher estimated cumulative reward than $\sigma_0$, then $P_\psi[\sigma_1 \succ \sigma_0]$ will be closer to 1.
- This formula resembles a **softmax over cumulative rewards** — commonly used in training the reward model via **cross-entropy loss** from binary preference labels.

---

### Example

Suppose a human is shown two short trajectories from a stock trading agent:

- **$\sigma_0$ (Bad Trade):**
  - $(s_0, a_0)$: buys a stock at high price
  - $(s_1, a_1)$: sells at loss
- **$\sigma_1$ (Good Trade):**
  - $(s'_0, a'_0)$: buys at low price
  - $(s'_1, a'_1)$: sells at profit

Assume the learned reward model gives:

- $\hat{r}_\psi(s^0_0, a^0_0) = -0.5$, $\hat{r}_\psi(s^0_1, a^0_1) = -1.0$ → total: $-1.5$
- $\hat{r}_\psi(s^1_0, a^1_0) = 1.0$, $\hat{r}_\psi(s^1_1, a^1_1) = 1.5$ → total: $2.5$

Then the preference probability is:

$$
P_\psi[\sigma_1 \succ \sigma_0] = \frac{\exp(2.5)}{\exp(2.5) + \exp(-1.5)} \approx \frac{12.18}{12.18 + 0.22} \approx 0.982
$$

This means the model predicts there's a 98.2% chance a human would prefer $\sigma_1$ over $\sigma_0$.

---

### Use in Learning

This probability is used in a **loss function** such as binary cross-entropy to **train the reward model** $\hat{r}_\psi$ using human-labeled preference data over trajectory pairs. This is a key step in algorithms like **Deep RL from Human Preferences** or **RLHF (Reinforcement Learning from Human Feedback)**.

---


Why EXP?

1. The exponential function turns any number into a positive value, which is necessary because probabilities must be positive.
2. Exponentials grow quickly. This means if one trajectory has a slightly higher reward than another, the exponential will make that difference much more noticeable.

**Softmax Function**

The softmax function takes a vector of real-valued scores and converts them into probabilities:

$$
\text{softmax}(z_i) = \frac{\exp(z_i)}{\sum_{j=1}^{n} \exp(z_j)}
$$

- $z_i$: The $i$-th input score.
- $n$: Total number of scores.
- $\exp(z_i)$: Exponential function applied to the input score.
- The denominator ensures all outputs sum to 1, forming a probability distribution.

---

#### Example:
Given scores $z = [2, 1, 0]$:

- $\exp(2) = 7.39$
- $\exp(1) = 2.72$
- $\exp(0) = 1$

Then:

$$
\text{softmax}(2) = \frac{7.39}{7.39 + 2.72 + 1} \approx 0.66 \\
\text{softmax}(1) \approx 0.24 \\
\text{softmax}(0) \approx 0.09
$$

So the softmax turns raw scores into a **normalized probability distribution** emphasizing the highest values.


**Preference Loss Function**

The loss function:

$$
L_{\text{pref}} = -\mathbb{E}_{(\sigma, y) \sim \mathcal{D}} \left[ y_0 \log P_\psi[\sigma_0 \succ \sigma_1] + y_1 \log P_\psi[\sigma_1 \succ \sigma_0] \right]
$$

is used in **preference-based reward learning**, where the goal is to train a reward model $\hat{r}_\psi$ from **human feedback** over pairs of trajectory segments.

---
**Plain English Explanation**

> “For each pair of trajectories $(\sigma_0, \sigma_1)$ and a human preference label $y$, this loss penalizes the model when its predicted preference probabilities disagree with the human's choice.”

---

**Term Breakdown**

- $\mathcal{D}$: The dataset of trajectory pairs with human preference labels.
- $(\sigma, y)$: A single data point, where:
  - $\sigma = (\sigma_0, \sigma_1)$ are two trajectory segments.
  - $y = (y_0, y_1)$ is a one-hot label:
    - $y_0 = 1$ if the human preferred $\sigma_0$.
    - $y_1 = 1$ if the human preferred $\sigma_1$.
- $P_\psi[\sigma_i \succ \sigma_j]$: The model's predicted probability (via softmax) that trajectory $\sigma_i$ is better than $\sigma_j$.
- The expression uses **log-probabilities** and is averaged across the dataset (expectation).
- The **negative sign** means this is a **loss to be minimized**, similar to cross-entropy loss.

---

**Example**

If the human prefers $\sigma_1$, then $y = (0, 1)$. The loss becomes:

$$
L_{\text{pref}} = -\log P_\psi[\sigma_1 \succ \sigma_0]
$$

So the model is penalized if it assigns a **low probability** to the trajectory that the human **actually preferred**.

---

Purpose **bold text**

This loss trains the reward model $\hat{r}_\psi$ to produce scores that match human preferences, which is essential in systems like **RLHF (Reinforcement Learning from Human Feedback)**.

**🔍 Saliency Maps and SmoothGrad Explained**

**What is a Saliency Map?**

A **saliency map** is a visualization that shows **which parts of an input are most influential** in a neural network’s decision. It answers the question:

> *“Which parts of this input does the model rely on most to make its prediction?”*

**📸 Example (Image Classification)**
For an image classifier that predicts "cat" or "dog":
- A saliency map highlights the most important pixels (e.g., cat’s ears or eyes) that influenced the model's decision.
- It is computed by taking the **gradient of the output class score with respect to each input pixel**:
  
  $$
  \nabla_x f_\theta(x)
  $$

  where:
  - $f_\theta(x)$ is the model’s output
  - $x$ is the input
  - $\nabla_x$ denotes the gradient with respect to $x$

**✍️ Example (Text Classification)**
For a sentiment classifier, a saliency map might highlight important words like *"love"* or *"hate"* that influenced the sentiment score.

---

**🚧 Problem: Raw Gradients Are Noisy**

Saliency maps computed directly from gradients often look **noisy and unstable**, especially for deep neural networks. This makes them hard to interpret.

---

**✅ Solution: SmoothGrad**

**SmoothGrad** improves saliency maps by averaging gradients over multiple noisy versions of the input.

**📐 SmoothGrad Formula**

$$
S_{\text{SmoothGrad}} = \frac{1}{N_{\text{smooth}}} \sum_{i=1}^{N_{\text{smooth}}} \nabla_x f_\theta(x + \epsilon_i)
$$

- $x$: the original input
- $\epsilon_i$: small random noise (usually Gaussian)
- $f_\theta$: the model
- $\nabla_x f_\theta(x + \epsilon_i)$: the gradient (saliency map) for a noisy input
- $N_{\text{smooth}}$: number of noisy samples

**🔍 Intuition**

> Add noise to the input several times, compute a saliency map for each, and average them to get a smoother, more reliable explanation.

This helps **reduce noise** and **highlight stable patterns** the model consistently pays attention to.

---

**🧠 Why It Matters**

Saliency maps and SmoothGrad are important tools for:
- **Interpreting neural networks**
- **Debugging model behavior**
- **Increasing trust** in model predictions
- **Detecting bias** in how models make decisions

---
**✅ Solution: IntegratedGradients**

 IntegratedGradients  calculate multiple ex-
planations with respect to a baseline and average the results
to increase stability.

---
**Other Solution**

Saliency-based training methods, such as [Expected-
Gradients](https://openreview.net/pdf?id=rygPm64tDH) and [Right-for-the-Right-Reasons](https://openreview.net/pdf?id=rygPm64tDH), allow the shaping of the training dynamics of neural networks.

How to simulate wind?
- Previous studies have often used <u>Gaussian noise</u> to simulate the uncertainty of natural
wind, such as the study by Zheng et al., which described the stochastic time-varying
characteristics of natural wind using sinusoidal Gaussian noise.
  - A better one: The Ornstein–Uhlenbeck (OU) process

**Ornstein–Uhlenbeck (OU) process**


**🔢 Equation**

$$
dx_t = \theta(\mu - x_t)dt + \sigma dW_t
$$

> "dee x sub t equals theta times the quantity mu minus x sub t, dee t, plus sigma dee W sub t."

>The small change in $x$ over time is equal to: a strength factor, called $\theta$, times how far we are from the target value $\mu$, **plus** some random noise that’s scaled by a noise level $\sigma$.
---

**🧠 Intuition**

A value that gets pulled toward a long-term average (mean reversion), but is also affected by random fluctuations (noise).

---

**📌 Terms Explained**

| Symbol   | Name                      | Meaning                                      |
|----------|---------------------------|----------------------------------------------|
| `x_t`    | Process value at time `t` | The current value of the system              |
| `μ`      | Long-term mean            | The value the system tends to revert to      |
| `θ`      | Reversion rate            | How strongly the system is pulled back to μ  |
| `σ`      | Volatility                | How strong the randomness is                 |
| `dt`     | Time increment            | A small step in time                         |
| `dW_t`   | Wiener process increment  | A small random step from Brownian motion     |

---

**🔍 Term-by-Term Meaning**

- `θ(μ - x_t)dt`:  
  The **mean-reverting force**. Pulls the value toward `μ`. Stronger when far from the mean.

- `σ dW_t`:  
  The **random shock**. Adds noise based on a Wiener process.

---

**🔬 What is a Wiener Process?**

- A continuous-time random walk.
- Each increment `dW_t` is normally distributed:  
  `dW_t ~ N(0, dt)`
- Models unpredictable, noisy behavior.

---

**🔁 Behavior Over Time**

- The process fluctuates randomly.
- But it **reverts** toward the mean `μ` over time.
- Unlike Brownian motion, it does not drift without bound.




```python
import numpy as np
import matplotlib.pyplot as plt

# OU Process parameters
theta = 0.5   # rate of mean reversion
mu = 1.0      # long-term mean
sigma = 0.3   # volatility
dt = 1.0      # time step
T = 50        # total number of steps
x0 = 2.0      # initial value

# Simulate the OU process
np.random.seed(42)  # for reproducibility
x = np.zeros(T)
x[0] = x0
for t in range(1, T):
    dx = theta * (mu - x[t-1]) * dt + sigma * np.random.normal() * np.sqrt(dt)
    x[t] = x[t-1] + dx

# Plot the result
plt.figure(figsize=(10, 5))
plt.plot(range(T), x, marker='o')
plt.axhline(mu, color='r', linestyle='--', label='Mean (μ = 1.0)')
plt.title('Ornstein–Uhlenbeck Process Simulation')
plt.xlabel('Time Step')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**autocorrelation function (ACF) for wind speed data**

if the time series represents wind speed data, the high autocorrelation
values indicate that wind speed at a given time is strongly correlated with wind speeds at previous times.


>If something is **statistically significant**, it means it’s strong enough that we believe it’s real, not just a random fluke.



**📊 How the Y-Values in the Autocorrelation Plot Are Computed**

The y-axis in the autocorrelation plot represents the **autocorrelation coefficients** at different time lags. These values measure how strongly the time series is correlated with itself after being shifted by a certain number of steps.

**Formula**

The autocorrelation at lag $k$ is given by:

$$
\rho_k = \frac{\sum_{t=1}^{n-k} (x_t - \bar{x})(x_{t+k} - \bar{x})}{\sum_{t=1}^{n} (x_t - \bar{x})^2}
$$

Where:

- $\rho_k$: autocorrelation at lag $k$  
- $x_t$: the value of the time series at time $t$  
- $\bar{x}$: the mean of the series  
- $n$: the total number of observations

**🧠 Intuition**

- At **lag 1**, you compare the series to itself shifted by one time step.
- A value close to **1** means strong positive correlation (values stay similar).
- A value near **0** means no correlation.
- A value near **−1** indicates strong negative correlation (values go in opposite directions).

Autocorrelation helps detect **temporal dependencies** in time series data, and the Ornstein–Uhlenbeck process typically shows **high positive autocorrelation** that decays slowly over time.

> **Shaded Blue (95\% Confidence Interval).** If the time series has no autocorrelation, the sample autocorrelation at any lag should be close to zero. But due to randomness, it may fluctuate a little. The confidence interval tells us how much fluctuation is normal under the assumption of no autocorrelation.

**How the 95% Confidence Interval is Computed in Autocorrelation Plots**

In autocorrelation plots, the shaded area typically shows the **95% confidence interval** under the assumption that the time series is **white noise** (i.e., purely random with no true autocorrelation).

**Formula for the Confidence Interval**

For large samples, the 95% confidence interval is approximately:

$$
\pm \frac{1.96}{\sqrt{n}}
$$

Where:

- $n$ is the sample size (number of observations in the time series)  
- $1.96$ is the critical value from the standard normal distribution corresponding to the 95% confidence level

**Interpretation**

- If an autocorrelation value falls **within** this interval, it is likely due to **random chance**.  
- If it falls **outside** the interval, it is considered **statistically significant** — suggesting a real correlation, not just noise.

**In Practice**

Tools like `statsmodels.plot_acf()` automatically compute and display this interval.  
The blue shaded region in the plot is this 95% range.  
Bars outside of this region indicate lags with statistically significant autocorrelation.
